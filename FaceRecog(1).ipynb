```python
#Importing basic libraries
import cv2
import numpy as np

# Load HAAR face classifier
face_classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

# Load function to extract face
def face_extractor(img):
    #Converts to gray scale
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    #Used to find find faces or eyes in the frame
    faces = face_classifier.detectMultiScale(gray, 1.3, 5)
    
    if faces is ():
        #Returns input image if face not found
        return None
    
    # Crop all faces found
    # If face is found, it returns the positions of detected faces as Rect(x,y,w,h)
    # (x, y), (x+w, y+h) are the four pixel locations for detected face
    #2 is the thickness of the line measured in pixels
    #(0, 255, 0) is the color of the shape, here green
    for (x,y,w,h) in faces:
        cropped_face = img[y:y+h, x:x+w]

    return cropped_face

# Initialize Webcam
cap = cv2.VideoCapture(0)
count = 0

# Collect 100 samples of your face from webcam input as database 
# This one-person database will be used as Biometrics for facial recognition of that person only (recognizes only the registered person)
while True:

    ret, frame = cap.read()
     # If face is detected from the frame it is resized and convered to gray scale
    if face_extractor(frame) is not None:
        count += 1
        face = cv2.resize(face_extractor(frame), (200, 200))
        face = cv2.cvtColor(face, cv2.COLOR_BGR2GRAY)

        # Sample collected is stored in a respective directory as Training Data
        # Files will be saved as 1.jpg, 2.jpg and so on..
        file_name_path = 'C:\\Users\\KIIT\\Desktop\\faceRecog\\Lupita\\' + str(count) + '.JPG'
        cv2.imwrite(file_name_path, face)

        # Display live count during sample collection
        cv2.putText(face, str(count), (50, 50), cv2.FONT_HERSHEY_COMPLEX, 1, (0,255,0), 2)
        cv2.imshow('Face Cropper', face)
        print("Face found")
        
    else:
        print("Face not found")
        pass

    if cv2.waitKey(1) == 13 or count == 100: #13 is the Enter Key
        break
        
cap.release()
cv2.destroyAllWindows()      
print("Collecting Samples Complete")
```

```python
import cv2
import numpy as np
from os import listdir
from os.path import isfile, join
print(cv2.__version__)

# Get the training data we previously made from the respective directory
data_path = 'C:\\Users\\KIIT\Desktop\\faceRecog\\Lupita\\'
# a=listdir('d:/faces')
# print(a)
# """

# onlyfiles consists of files for the database
onlyfiles = [f for f in listdir(data_path) if isfile(join(data_path, f))]

# Create numpy arrays for training data and labels
Training_Data, Labels = [], []

# (1)Open training images from our database
# (2)Reading the images in Grey scale and appending them to their labels
for i, files in enumerate(onlyfiles):
    image_path = data_path + onlyfiles[i]
    images = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    Training_Data.append(np.asarray(images, dtype=np.uint8))
    Labels.append(i)
# 
# Create a numpy array for both training data and labels
Labels = np.asarray(Labels, dtype=np.int32)
#Facial Recognizer
model=cv2.face_LBPHFaceRecognizer.create()

# Let's train our model after Training data and Labels have been convered to numpy arrays and Training Data is appended with 
# their respective labels
model.train(np.asarray(Training_Data), np.asarray(Labels))
print("Model trained sucessefully")

```

```python
import cv2
import numpy as np
import webbrowser

# For Face Detection from the video frame
face_classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

def face_detector(img, size=0.5):
    
    # Convert live image to grayscale
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces = face_classifier.detectMultiScale(gray, 1.3, 5)
    if faces is ():
        return img, []
    
    #Creates bounding boxes if image is found in frame and resizes the image (roi)
    for (x,y,w,h) in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,255),2)
        roi = img[y:y+h, x:x+w]
        roi = cv2.resize(roi, (200, 200))
    return img, roi


# Open Webcam
cap = cv2.VideoCapture(0)

while True:

    ret, frame = cap.read()
    
    image, face = face_detector(frame)
    
    try:
        face = cv2.cvtColor(face, cv2.COLOR_BGR2GRAY)

        # Pass face to prediction model
        # "results" comprises of a tuple containing the face label and the confidence value of recognition from the face recognized in the frame
        results = model.predict(face)
        print(results)
        if results[1] < 500: #500 is threshold for user confidence
            confidence = int( 100 * (1 - (results[1])/400) )
            display_string = str(confidence) + '% Confident it is User'
            
        cv2.putText(image, display_string, (100, 120), cv2.FONT_HERSHEY_COMPLEX, 1, (255,120,150), 2)
        
        if confidence > 75: #75 is threshold for confidence of the registered person
            cv2.putText(image, "Hey Lupita ! ", (250, 450), cv2.FONT_HERSHEY_COMPLEX, 1, (0,255,0), 2)
            cv2.imshow('Face Recognition', image )
            #break
        else:
            cv2.putText(image, "Locked", (250, 450), cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,255), 2)
            cv2.imshow('Face Recognition', image )

    except:
        cv2.putText(image, "No Face Found", (220, 120) , cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,255), 2)
        cv2.putText(image, "Locked", (250, 450), cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,255), 2)
        cv2.imshow('Face Recognition', image )
        pass
        
    if cv2.waitKey(1) == 13: #13 is the Enter Key
        break
        
cap.release()
cv2.destroyAllWindows()   
```
